```python
import numpy as np                                                  # 导入模块
import pandas as pd
import scipy.optimize as spo
import matplotlib.pyplot as plt
import time
XH_pd = pd.read_csv('E:/xxx/SEM/XH.csv', nrows =1006, header=None)  # 读取数据
XL_pd = pd.read_csv('E:/xxx/SEM/XL.csv', nrows =1006, header=None)
YH_pd = pd.read_csv('E:/xxx/SEM/YH.csv', nrows =1006, header=None)
YL_pd = pd.read_csv('E:/xxx/SEM/YL.csv', nrows =1006, header=None)
XH = XH_pd.values                                                   # 矩阵格式
XL = XL_pd.values
YH = YH_pd.values
YL = YL_pd.values
Xposition=(XH*65536 + XL)*0.0000386;                                # 计算坐标
Yposition=(YH*65536 + YL)*0.0000386;
Xposition_pd = pd.DataFrame(Xposition)                              # pands格式
Yposition_pd = pd.DataFrame(Yposition)
Xposition_pd.to_csv('E:/xxx/SEM/Xposition.csv')                     # 写入文件
Yposition_pd.to_csv('E:/xxx/SEM/Yposition.csv')
```
